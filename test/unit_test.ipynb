```python
import subprocess

s = subprocess.check_output("g++ unit_test.cpp generator.cpp generator.h -o out;./out" , shell = True, cwd='/home/jctuesta/Documents/cppPractice/python/Knapsack_problem/scr/')

a = list(s.decode("utf-8").split())
a = [int(x) for x in a]
size_input = len(a)
```

```python
weight = a[0:int(size_input/2)]
profit = a[int(size_input/2):]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.plot(range(0, len(weight)), weight, label = "weight")
plt.plot(range(0, len(profit)), profit, label = "profit")
plt.xlabel('elements')
plt.ylabel('Weight')
plt.legend()

plt.show()
```
